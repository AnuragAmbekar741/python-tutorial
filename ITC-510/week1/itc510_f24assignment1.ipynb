### 1. In the Code cell below, write a function that calculates the distance between of two **integer** points in a plane using the coordinates (x1, y1) and (x2, y2) where the values of the two points are entered by the user and then passed to the function individually as four separate  values.  

### Include your test case code which should contain several calls to your function some of which return successfully and some that return error messages when the function is given invalid input.


```python
import math

def getDistance(x1, y1, x2, y2):
    try:
        x1 = int(x1)
        y1 = int(y1)
        x2 = int(x2)
        y2 = int(y2)
        
        distance = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
        return distance 
    
    except ValueError:
        return "Error: Invalid input. Please enter valid integers."

# Test cases
print(getDistance(3.3, 4.2, 7.1, 1.1))  # Valid input
print(getDistance(-1, -1, 2, 3))  # Valid input
print(getDistance("a", 5, 6, 7))  # Invalid input (non-integer)
print(getDistance(2, 3, "x", "y"))  # Invalid input (non-integer)
print(getDistance(0, 0, 0, 0))  # Valid input (same point)
```

### 2.  In the code cell below, write a function that calculates the summation (the sum of the sequence) between two **positive** integers.  The two integers must be passed to the function in a single two-element Python list.  If provided valid input, the function should calculate the summation and return the result of the calculation.

### Include your test case code.


```python
def getSumFromList(num_list):
    try:

        if(type(num_list)!=list):
            return "Error: Input must be a list."


        if len(num_list) != 2:
            return "Error: Input must be a list of exactly two integers."
        
        num1 = int(num_list[0])
        num2 = int(num_list[1])
        
        if num1 <= 0 or num2 <= 0:
            return "Error: Both numbers must be positive integers."

        total_sum = 0
        
        if num1 <= num2:
            for i in range(num1, num2 + 1):
                total_sum += i
        else:
            for i in range(num2, num1 + 1):
                total_sum += i
        
        return total_sum
    
    except ValueError:
        return "Error: Input must contain valid integers."


```

```python
print(getSumFromList([3,2]))  
print(getSumFromList([15, 10]))  
print(getSumFromList([3, -1]))  
print(getSumFromList([3]))      
print(getSumFromList([3, 'b']))
print(getSumFromList(3))
```

### For this problem, you must first create a text file called "words.txt" that contains at least 100 words.   Then prior to calling your function you must first read in the "words.txt" file and place each "word" in a list that is then passed to passed to your function.

```python
import os

#method/function to read file.
def readWordsFromFile(filename):
    if not os.path.exists(filename):
        return (f"The file '{filename}' does not exist.")
    
    with open(filename, 'r') as file:
        words = [word.strip() for word in file.readlines()]
        
    return words

#function/method for checking word couut and making dist and returning list
def filterWordsAsPerOcc(word_list, target_count):
    if not isinstance(word_list, list):
        raise ValueError("The first argument must be a list.")
    if not all(isinstance(word, str) for word in word_list):
        raise ValueError("The list must contain only strings.")
    if not isinstance(target_count, int) or target_count <= 0:
        raise ValueError("The second argument must be a positive integer.")
    
    word_count = {}
    for word in word_list:
        if word in word_count:
            word_count[word] += 1
        else:
            word_count[word] = 1
    
    result_list = [word for word, count in word_count.items() if count == target_count]
    
    return result_list
```

```python

# Read the list of words from the file
words = readWordsFromFile('dummy.txt')

result = filterWordsAsPerOcc(words, 2)
print(f"Words appearing exactly 2 times: {result}")

result = filterWordsAsPerOcc(words, 3)
print(f"Words appearing exactly 3 times: {result}")

words = readWordsFromFile('notpresent.txt')
result = filterWordsAsPerOcc(words, 3)
print(f"Words appearing exactly 3 times: {result}")
        
```
